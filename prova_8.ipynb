```python
import matplotlib
import matplotlib.pyplot as plt
import numpy as np

vendas_2018 = [2.0, 3.4, 3.0, 3.5]
vendas_2019 = [2.5, 3.2, 3.4, 2.0]
espessura = 0.35
rotulos = ['Q1', 'Q2', 'Q3', 'Q4']
x = np.arange(len(rotulos))

fig, ax = plt.subplots(figsize=(12, 8))

ax.bar(x - espessura/2, vendas_2018, espessura, label='2018', color='tab:red')
ax.bar(x + espessura/2, vendas_2019, espessura, label='2019', color='tab:green')

ax.set_title('Vendas por quadrimestre/ano (em milhões)')
ax.set_xlabel('Quadrimestre')
ax.set_ylabel('Milhões')
ax.set_xticks(x, rotulos)
ax.set_ylim(0, 3.6)
ax.legend(loc='upper right')

plt.tight_layout()
plt.show()
```
