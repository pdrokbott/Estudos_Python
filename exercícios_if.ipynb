Exercício Python 36: Escreva um programa para aprovar o empréstimo bancário para a compra de uma casa. Pergunte o valor da casa, o salário do comprador e em quantos anos ele vai pagar. A prestação mensal não pode exceder 30% do salário ou então o empréstimo será negado.

```python
valor = float(input('qual o valor da casa?' ))
salario = float(input('qual o seu salario?' ))
anos = float(input('quantos anos voce vai pagar?'))

meses = anos * 12
prestacao_mensal = valor / meses 

if prestacao_mensal <= (0.3 * salario):
    print('aprovado')
    print(f'sua prestacao é de {prestacao_mensal:.2f}')
    
else:
    print(f'você nao consegue sua prestação excede 30%')
    print(f'sua prestacao é de {prestacao_mensal:.2f}')
    print(meses)
    print(salario)


```

Exercício Python 37: Escreva um programa em Python que leia um número inteiro qualquer e peça para o usuário escolher qual será a base de conversão: 1 para binário, 2 para octal e 3 para hexadecimal.

```python
num = int(input('insira um numero que voce quer seja convertido'))
opcao = int(input(f'digite [1] para binário, [2] para octal, [3] para hexadecimal'))

if opcao == 1:
    con = bin(num)
    print(con)
elif opcao == 2:
    con = oct(num)
    print(con)
elif opcao == 3:
    con = hex(num)
    print(con)
else:
    print('opcao digitada incorreta')
```

– IMC abaixo de 18,5: Abaixo do Peso

– Entre 18,5 e 25: Peso Ideal

– 25 até 30: Sobrepeso

– 30 até 40: Obesidade

– Acima de 40: Obesidade Mórbida

Aula Anterior


```python
altura = float(input('digite a sua altura'))
peso = float(input('digite a seu peso'))
imc = peso/ (altura**2)

if imc < 18.5: 
    print(f'abaixo do peso {imc:.2f}')
elif imc <= 25:
    print(f'peso ideal {imc:.2f}')
elif imc <= 30:
    print(f'sobrepeso {imc:.2f}')
elif imc <= 40:
    print(f'obesidade {imc:.2f}')
else:
    print(f'Obesidade morbida {imc:.2f}')
```

Exercício Python 44: Elabore um programa que calcule o valor a ser pago por um produto, considerando o seu preço normal e condição de pagamento:

– à vista dinheiro/cheque: 10% de desconto

– à vista no cartão: 5% de desconto

– em até 2x no cartão: preço formal 

– 3x ou mais no cartão: 20% de juros

```python
preco = float(input('Coloque o preço do produto'))
pagamento = int(input('escreva [1] para dinheiro/cheque, [2] a vista no cartão, [3] parcelado em 2x, [4] 3x ou mais no cartão'))

if pagamento == 1:
    valor = preco * 0.9
    print(f'Você pagou no dinheiro com o desconto de 10% foi de {preco} para {valor}')
elif pagamento == 2:
    valor = preco * 0.95
    print(f'Você pagou a vista no cartao com o desconto de 5% foi de {preco} para {valor}')
elif pagamento == 3:
    print(f'Você pagou em 2x no cartao com o preço normal {preco}')
elif pagamento == 4:
    valor = preco * 1.2
    print(f'Você pagou em 3x ou mais no cartao com o 20% de juros foi de {preco} para {valor}')
else:
    print('tipo de pagamento errado')
```
