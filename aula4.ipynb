```python
notas = []

while True:
    nota = float(input("insira suas nota ou digite -1 para sair"))
    if nota == -1:
        break
    notas.append(nota)

    if len(notas) > 0:
        media = sum(notas)/len(notas)
        maior = max(notas)
        menor = min(notas)

        print(f"MÃ©dia: {media:.2f}")
        print(f"Maior nota: {maior}")
        print(f"Menor nota: {menor}")
    else:
        print("Nenhuma nota foi digitada.")
```
