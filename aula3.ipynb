## 1) Classificador de IMC com valida√ß√£o
Pe√ßa `peso` (kg) e `altura` (m).  
Valide:
- peso > 0 e altura > 0
- altura razo√°vel (ex.: 0.5 a 2.5)

Se inv√°lido ‚Üí mensagem de erro.  
Se v√°lido ‚Üí calcule IMC = peso / (altura**2) e classifique:

- < 18.5 ‚Üí Abaixo do peso  
- 18.5 a 24.9 ‚Üí Peso normal  
- 25.0 a 29.9 ‚Üí Sobrepeso  
- ‚â• 30.0 ‚Üí Obesidade

Exiba o IMC com 2 casas decimais e a categoria.

````python
peso = float(input("insira seu peso"))
altura = float(input("agora insira a sua altura"))
if (peso > 0) and (0.5 < altura < 2.5):
    IMC = peso / (altura**2)
    print(IMC)
    if IMC >= 30:
        print(f"voc√™ esta obeso baleia seu {IMC:.2f}")
    elif 25.0 <= IMC < 29.9:
        print(f"voc√™ esta com sobrepeso seu {IMC:.2f}")
    elif 18.5 <= IMC < 24.9:
        print(f"voc√™ esta com o IMC normal {IMC:.2f}")
    else:
        print("voc√™ esta abaixo do peso")


else: 
    print("peso ou altura invalidos")
````

## 2) Desconto progressivo com m√∫ltiplas regras
Entrada: `valor_compra`, `tipo_cliente` (`"vip"` ou `"comum"`), `cupom` (texto ou vazio).  
Regras de desconto (aplique **uma √∫nica** ‚Äî a melhor aplic√°vel):

- Se `tipo_cliente == "vip"` e `valor_compra >= 500` ‚Üí 12%  
- Sen√£o, se `valor_compra >= 400` ‚Üí 10%  
- Sen√£o, se `valor_compra >= 250` ‚Üí 7%  
- Sen√£o, se `cupom == "SALE10"` ‚Üí 10%  
- Caso contr√°rio ‚Üí 0%

Mostre:
- Valor original, desconto aplicado (% e R$), valor final.

Dica: cuide da **prioridade** (ordem das condi√ß√µes).

````python
valor_compra = float(input("Insira o valor da compra: "))
tipo_cliente = input("Qual o seu tipo de cliente (vip/comum)? ").lower().strip()
cupom = input("Insira o cupom (ou deixe vazio): ").upper().strip()

desconto = 0

if tipo_cliente == "vip" and valor_compra >= 500:
    desconto = 0.12
elif valor_compra >= 400:
    desconto = 0.10
elif valor_compra >= 250:
    desconto = 0.07
elif cupom == "SALE10":
    desconto = 0.10

valor_desconto = valor_compra * desconto
valor_final = valor_compra - valor_desconto

print(f"\nValor original: R$ {valor_compra:.2f}")
print(f"Desconto aplicado: {desconto*100:.0f}% (-R$ {valor_desconto:.2f})")
print(f"Valor final: R$ {valor_final:.2f}")
````

## üß© Exerc√≠cio 2 ‚Äî Controle de estoque
Pe√ßa o **estoque inicial** e repita enquanto ele for maior que 0.

- A cada itera√ß√£o, pe√ßa `quantidade_vendida`.
- Subtraia do estoque.
- Se a venda for maior que o estoque, mostre `"Venda inv√°lida"`.
- Quando zerar, mostre `"Estoque esgotado"`.

üí° *Dica:* use `continue` pra ignorar tentativas inv√°lidas.

````python
estoque = int(input("insira seu estoque inicial"))

while estoque > 0:
    quantidade_vendida = int(input("insira venda"))
    if quantidade_vendida > estoque:
        print("Venda Inv√°lida")
    else:
        estoque = estoque - quantidade_vendida
        print(f"estoque restante {estoque}")
    
if estoque <= 0:
    print("Estoque Esgotado")
````

## üß© Exerc√≠cio 3 ‚Äî C√°lculo progressivo
Pe√ßa notas (usando `float`) at√© o usu√°rio digitar `-1`.

Mostre:
- Quantas notas foram digitadas  
- A m√©dia  
- A maior e a menor nota  

Exemplo:

Exemplo:
Digite uma nota (-1 para sair): 7
Digite uma nota (-1 para sair): 8
Digite uma nota (-1 para sair): 5
Digite uma nota (-1 para sair): -1

M√©dia: 6.67
Maior nota: 8.0
Menor nota: 5.0

yaml
Copiar c√≥digo

üí° *Dica:* use uma lista (`notas = []`) ou vari√°veis `soma`, `contagem`, `maior`, `menor`.

````python
notas = []

while True:
    nota = float(input("Digite uma nota (-1 para sair): "))
    if nota == -1:
        break
    notas.append(nota)
    print(notas)

if len(notas) > 0:
    media = sum(notas) / len(notas)
    maior = max(notas)
    menor = min(notas)

    print(f"M√©dia: {media:.2f}")
    print(f"Maior nota: {maior}")
    print(f"Menor nota: {menor}")
else:
    print("Nenhuma nota foi digitada.")
````
